```python
# imports
import pandas as pd
```

```python
# configs
DATA_PATH = "../data/raw/all_tickets_processed_improved_v3.csv"
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
print("Unique classes: " + str(df["Topic_group"].nunique()))
print("\nClasswise count")
print(df["Topic_group"].value_counts())
print("\nDocement length")
df["Document"].str.len().describe()
```

```python
print("\nTop 2 classes")
print(df["Topic_group"].value_counts()[:2])
print("\nBottom 2 classes")
print(df["Topic_group"].value_counts()[-2:])
```

```python
from sklearn.model_selection import train_test_split

X = df["Document"]
y = df["Topic_group"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, shuffle=True, random_state=2)
```

```python
# train test dimentions
print(f"Raw: X shape: {X.shape} | y shape: {y.shape}")
print(f"Train: X shape: {X_train.shape} | y shape: {y_train.shape}")
print(f"Test: X shape: {X_test.shape} | y shape: {y_test.shape}")
```

```python
# tfidf test
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=30000,
    min_df=3
)

print(f"Test TF-IDF shape: {vectorizer.fit_transform(X_train[:10]).shape}")
vectorizer.get_feature_names_out()
```

```python
# tfidf transform - full data
vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=30000,
    min_df=3
)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print(f"TF-IDF: X train shape: {X_train_tfidf.shape} | X test shape: {X_test_tfidf.shape}")
```
